```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('img6.avif', 0)
_, binary = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)

binary_inverted = cv2.bitwise_not(binary)

kernel = np.ones((5, 5), np.uint8)

erosion = cv2.erode(binary_inverted, kernel, iterations=1)
dilation = cv2.dilate(binary_inverted, kernel, iterations=1)
opening = cv2.morphologyEx(binary_inverted, cv2.MORPH_OPEN, kernel)
closing = cv2.morphologyEx(binary_inverted, cv2.MORPH_CLOSE, kernel)

plt.figure(figsize=(15, 10))
images = [binary_inverted, erosion, dilation, opening, closing]
titles = ['Original (Inverted)', 'Erosion', 'Dilation', 'Opening', 'Closing']

for i in range(5):
    plt.subplot(2, 3, i+1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
